### Read selected columnsm

### Extract records based on single condition on table

```python
query = """
SELECT *
FROM countrylanguage
WHERE Isofficial = 'T'
"""

data = pd.read_sql_query(query, myconn)

data
```

### Extract records based on mathematical condition or column

```python
query = """
SELECT *
FROM countrylanguage
WHERE Percentage > 75
"""

data = pd.read_sql_query(query, myconn)

data
```

### Extract data based on two conditions

```python
# Extract records from table 'countrylanguage' with Percentage > 75 and Isofficial status is 'T'

query = """
SELECT *
FROM countrylanguage
WHERE Percentage > 75
AND Isofficial = 'T'
"""

data = pd.read_sql_query(query, myconn)

data
```

```python
# Extract records from table 'countrylanguage' with Percentage > 75 or Isofficial status is 'T'

query = """
SELECT *
FROM countrylanguage
WHERE Percentage > 75
OR Isofficial = 'T'
"""

data = pd.read_sql_query(query, myconn)

data
```

### Get Result in sorted form

### Read data from table 'city' with Name is ascending order and 'Population' in descending order.
